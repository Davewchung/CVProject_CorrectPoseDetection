```python
import mediapipe as mp
import numpy as np
import statistics


mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_pose = mp.solutions.pose
downdog_joints_xy = [[] for _ in range(66)]
goddess_joints_xy= [[] for _ in range(66)]
plank_joints_xy = [[] for _ in range(66)]
tree_joints_xy = [[] for _ in range(66)]
warrior2_joints_xy = [[] for _ in range(66)]

downdog_joints_angle = {}
goddess_joints_angle= {}
plank_joints_angle = {}
tree_joints_angle = {}
warrior2_joints_angle = {}
```

```python
def calculate_angle(a,b,c): 
  a = np.array(a)
  b = np.array(b)
  c = np.array(c)

  radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
  angle = np.abs(radians*180/np.pi)

  if angle > 180.0: 
    angle = 360-angle

  return angle
```

```python
def get_angles(xy_list, angle_dicts):
  left_shoulder =  [xy_list[22], xy_list[23]]
  right_shoulder = [xy_list[24], xy_list[25]]
  left_elbow = [xy_list[26], xy_list[27]]
  right_elbow = [xy_list[28], xy_list[29]]
  left_wrist = [xy_list[30], xy_list[31]]
  right_wrist = [xy_list[32], xy_list[33]]
  left_index = [xy_list[38], xy_list[39]]
  right_index = [xy_list[40], xy_list[41]]
  left_hip = [xy_list[46], xy_list[47]]
  right_hip = [xy_list[48], xy_list[49]]
  left_knee = [xy_list[50], xy_list[51]]
  right_knee = [xy_list[52], xy_list[53]]
  left_ankle = [xy_list[54], xy_list[55]]
  right_ankle = [xy_list[56], xy_list[57]]
  left_foot_index = [xy_list[62], xy_list[63]]
  right_foot_index = [xy_list[64], xy_list[65]]

  LShoulder_angle = calculate_angle(right_shoulder, left_shoulder, left_elbow) #LShoulder_angle
  RShoulder_angle = calculate_angle(left_shoulder, right_shoulder, right_elbow) #RShoulder_angle
  LElbow_angle= calculate_angle(left_shoulder, left_elbow, left_wrist) #LElbow_angle
  RElbow_angle = calculate_angle(right_shoulder, right_elbow, right_wrist) #RElbow_angle
  LWrist_angle = calculate_angle(left_elbow, left_wrist, left_index) #LWrist_angle
  RWrist_angle = calculate_angle(right_elbow, right_wrist, right_index) #RWrist_angle
  LHip_angle = calculate_angle(left_shoulder, left_hip, left_knee) #LHip_angle
  RHip_angle = calculate_angle(right_shoulder, right_hip, right_knee) #RHip_angle
  LKnee_angle = calculate_angle(left_hip, left_knee, left_ankle) #LKnee_angle
  RKnee_angle = calculate_angle(right_hip, right_knee, right_ankle) #RKnee_angle

  if LKnee_angle > RKnee_angle: 
      moreBentKnee = RKnee_angle
      lessBentKnee = LKnee_angle
  else: 
      moreBentKnee = LKnee_angle
      lessBentKnee = RKnee_angle

  LAnkle_angle = calculate_angle(left_knee, left_ankle, left_foot_index) #LAnkle_angle
  RAnkle_angle  = calculate_angle(right_knee, right_ankle, right_foot_index) #RAnkle_angle

  if LHip_angle > RHip_angle: 
      moreBentHip = RHip_angle
      lessBentHip = LHip_angle
  else: 
      moreBentHip = LHip_angle
      lessBentHip = RHip_angle


  if LWrist_angle > RWrist_angle: 
      moreBentWrist = RWrist_angle
      lessBentWrist = LWrist_angle
  else: 
      moreBentWrist = LWrist_angle
      lessBentWrist = RWrist_angle

  LAnkle_angle = calculate_angle(left_knee, left_ankle, left_foot_index) #LAnkle_angle
  RAnkle_angle  = calculate_angle(right_knee, right_ankle, right_foot_index) #RAnkle_angle

  angle_dicts['LShoulder'].append(LShoulder_angle)
  angle_dicts['RShoulder'].append(RShoulder_angle)
  angle_dicts['LElbow'].append(LElbow_angle)
  angle_dicts['RElbow'].append(RElbow_angle)
  angle_dicts['moreBentKnee'].append(moreBentKnee)
  angle_dicts['lessBentKnee'].append(lessBentKnee)
  angle_dicts['moreBentHip'].append(moreBentHip)
  angle_dicts['lessBentHip'].append(lessBentHip)
  angle_dicts['lessBentWrist'].append(lessBentWrist)
  angle_dicts['moreBentWrist'].append(moreBentWrist)
  angle_dicts['LAnkle'].append(LAnkle_angle)
  angle_dicts['RAnkle'].append(RAnkle_angle)
  return angle_dicts

```

```python
def joints_xy(IMAGE_FILES, joints_xy): 
  BG_COLOR = (192, 192, 192) # gray 
  with mp_pose.Pose(
      static_image_mode=True,
      model_complexity=2,
      enable_segmentation=True,
      min_detection_confidence=0.5) as pose:
    for idx,image in enumerate(IMAGE_FILES):
      # Convert the BGR image to RGB before processing.
      results = pose.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
      if not results.pose_landmarks:
        continue
      for i in range(0,33): 
        joints_xy[2*i].append(results.pose_landmarks.landmark[i].x)
        joints_xy[2*i+1].append(results.pose_landmarks.landmark[i].y)
```
